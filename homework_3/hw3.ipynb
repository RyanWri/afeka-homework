```python
# Import necessary libraries
import numpy as np
from question1 import plot_attention_weights, self_attention, multi_head_attention
```

```python
# Define polysemy examples
sentence1 = "The lead actor delivered an incredible performance."
sentence2 = "Exposure to lead is harmful to health."

tokens_sentence1 = sentence1.split()
tokens_sentence2 = sentence2.split()

# Mock embeddings
np.random.seed(42)
embeddings_sentence1 = np.random.rand(1, len(tokens_sentence1), 8)  # d_model = 8
embeddings_sentence2 = np.random.rand(1, len(tokens_sentence2), 8)
```

# 1A. Self Attention

```python
# Apply self_attention
output_sentence1, weights_sentence1 = self_attention(
    embeddings_sentence1, embeddings_sentence1, embeddings_sentence1
)
output_sentence2, weights_sentence2 = self_attention(
    embeddings_sentence2, embeddings_sentence2, embeddings_sentence2
)

# Visualize attention weights
plot_attention_weights(
    weights_sentence1[0], tokens_sentence1, "Self-Attention Weights: Sentence 1"
)
plot_attention_weights(
    weights_sentence2[0], tokens_sentence2, "Self-Attention Weights: Sentence 2"
)
```

# 1B.  Multi Head Attention

```python
# Apply multi_head_attention
multi_head_output1, multi_head_weights1 = multi_head_attention(embeddings_sentence1)
multi_head_output2, multi_head_weights2 = multi_head_attention(embeddings_sentence2)

# Visualize attention weights for each head
for i in range(multi_head_weights1.shape[0]):  # Iterate over heads
    plot_attention_weights(
        multi_head_weights1[i][0],
        tokens_sentence1,
        f"Head {i + 1} Attention: Sentence 1",
    )

for i in range(multi_head_weights2.shape[0]):  # Iterate over heads
    plot_attention_weights(
        multi_head_weights2[i][0],
        tokens_sentence2,
        f"Head {i + 1} Attention: Sentence 2",
    )
```
