**Observation**:

#### Lemmatization:
- **NLTK Unique Lemmas**: 2,379

- **Common Lemmas**: 8,221
- **Total Lemmas**:
  - NLTK: 104,904
  - SpaCy: 104,367

**Observation**:
- SpaCy's lemmatization is more conservative, producing fewer unique lemmas, while NLTK is more aggressive and often generates more variations.

#### Stemming:
- **NLTK Unique Stems**: 5,870
- **SpaCy Unique Stems**: 4,777
- **Common Stems**: 2,197
- **Total Stems**:
  - NLTK: 104,904
  - SpaCy: 104,367

**Observation**:
- NLTK's stemming is more aggressive, leading to more fragmented stems. SpaCy’s stems (approximated using `.lemma_`) are closer to lemmatization and thus less varied.

## Section 11. Comparison of NLTK Before and After Tokenization, Lemmatization, and Stemming

#### Before Tokenization:
- Messages are raw text with varying formats (e.g., punctuation, special characters, etc.).

#### After Tokenization:
- **Total Tokens**: 104,904
- **Unique Tokens**: 689
- NLTK tokenization splits text into words but does not handle linguistic nuances (e.g., contractions are split into separate tokens).

#### After Lemmatization:
- **Total Lemmas**: 104,904
- **Unique Lemmas**: 2,379
- Lemmatization reduces tokens to their dictionary forms but keeps linguistic variations (e.g., "better" → "good").

#### After Stemming:
- **Total Stems**: 104,904
- **Unique Stems**: 5,870
- Stemming is more aggressive, often stripping suffixes and creating non-dictionary words (e.g., "running" → "run", "better" → "better").

**Observation**:
- Tokenization introduces a structured list of words.
- Lemmatization normalizes words to base forms, leading to fewer unique entries.
- Stemming is more aggressive and results in more unique variations than lemmatization.

## section 12. Comparison of SpaCy Before and After Tokenization, Lemmatization, and Stemming

#### Before Tokenization:
- Messages are raw text, similar to the NLTK input.

#### After Tokenization:
- **Total Tokens**: 104,367
- **Unique Tokens**: 743
- SpaCy handles linguistic details better, preserving contractions (e.g., "I'm" remains a single token).

#### After Lemmatization:
- **Total Lemmas**: 104,367
- **Unique Lemmas**: 1,434
- SpaCy produces fewer unique lemmas, as it normalizes words more consistently to their dictionary forms (e.g., "better" → "good", "running" → "run").

#### After Stemming (Approximated):
- **Total Stems**: 104,367
- **Unique Stems**: 4,777
- SpaCy's stemming (approximated using `.lemma_[:n]`) is closer to its lemmatization, resulting in fewer unique variations.

**Observation**:
- Tokenization splits text more precisely, maintaining contractions as single tokens.
- Lemmatization reduces redundancy by normalizing words effectively.
- Stemming (if approximated) produces results closer to lemmatization compared to NLTK.
